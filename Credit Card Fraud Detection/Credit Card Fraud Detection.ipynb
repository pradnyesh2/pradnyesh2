# **Credit Card Fraud Detection**

Import all necessary liabraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score

from pylab import rcParams
rcParams['figure.figsize'] = 8,8
```

```python
data = pd.read_csv('creditcard.csv')
```

#### Things to do when we have imbalanced data:
1. Oversampling (SMOTE Technique)
2. Undersampling(Loss of Data)
3. Use Tree Based models only (Decision Tree, Random Forest, XGBoost, Adaboost, GradientBoost).

```python
# Now let's divide the input and output data

X = data.drop('Target', axis=1)
y = data['Target']
```

* Split the dataset into train and test using stratified sampling on our dependent variable. 
* Using a stratified sampling ensures the distribution of dependent variable remains same across train and test datasets

```python
# Split the dataset into Training and Testing

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify= y, random_state=42)
```

The stratify parameter asks whether you want to retain the same proportion of classes in the train and test sets that are found in the entire original dataset. It means if there are 60% 'yes' class data and 40% 'no' class data in the full dataset, then in train and test data split this proportion should be maintained.

## Apply ML to the Data

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
param_grid = { 
    'n_estimators': [25, 50, 100, 150], 
    'max_depth': [1,2,3,4,5,6,7,8,9], 
    'max_leaf_nodes': [3,6,9], 
}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(), param_grid=param_grid) 
grid_search.fit(X_train, y_train)
print(grid_search.best_estimator_)
```

```python
random_search = RandomizedSearchCV(RandomForestClassifier(), param_distributions=param_grid)
random_search.fit(X_train, y_train)
print(random_search.best_estimator_)
```

### Let's build a model with max_depth=5, max_leaf_nodes=9, n_estimators=25

```python
rf_classifier_1 = RandomForestClassifier(n_estimators=25, max_depth=5, max_leaf_nodes=9)
rf_classifier_1.fit(X_train, y_train)
```

**n_estimator** is the number of trees and ideally the value of this paramater `should not be less than 100` and **max_depth** helps us control the length of each tree in Random Forest

```python
y_pred = rf_classifier_1.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
y_pred_train = rf_classifier_1.predict(X_train)
```

```python
accuracy_score(y_train, y_pred_train)
```

**Accuracy is not a good metric to evaluate our model since the dataset is heavily imbalanced. Thus we will use roc_auc score and confusion matrix**

```python
roc_auc_score(y_test, y_pred)
```

```python
roc_auc_score(y_train, y_pred_train)
```
